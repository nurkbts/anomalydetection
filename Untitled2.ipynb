#                              K-nn Classification

Verilerimize ilk uygulanan algoritma K-Nearest Neighbors(K-nn) algoritmasıdır. K-nn en basit makine öğrenme algoritmasıdır. Yeni bir veri noktasına dair bir tahmin yapmak için algoritma, eğitim verisetindeki "en yakın komşuları" içindeki en yakın veri noktalarını bulur.Öngörüde bulunmak istediğimiz noktaya en yakın eğitim verilerini dikkate alır.Bütün eğitim seti içindekiler ile arasındaki uzaklık bulunur. En yakın k örnek seçilir. Çoğunluk olan sınıfa ait olur.Anlaşılması kolay, hızlı ama büyük miktardaki verilerde yavaş çalışmaktadır.

2 önemli parametresi vardır. Birincisi, komşu sayısı olan k' dır.k tamsayı olmalıdır. k' nın en kompleks katsayısı 1' dir.  K komşular algoritmasının k parametresini seçmek için birçok yöntem mevcut olsa da, en yaygın olanı veri kümesindeki farklı k değerlerini denemek ve en etkili olanı seçmektir.İkincisi, noktalar arasındaki mesafenin naşıl ölçüleceğidir.


Aşağıdaki en yakın 3 komşuluğa ait değerleri göstermektedir. 

```python
#en yakın 3 komşuluğa baktık
clf=KNeighborsClassifier(n_neighbors=3)
clf.fit(data_train,y_train)
print("Orjinal Verinin Train set accuracy: {:.3f}".format(clf.score(data_train,y_train)))
print("Orjinal Verinin Test set accuracy: {:.3f}".format(clf.score(data_test,y_test)))
print("Orjinal Verinin Test2 set accuracy: {:.3f}".format(clf.score(data_test2,y_test2)))
clf.fit(data_train_normalized,y_train_normalized)
print("------")
print("Normalize Verinin Train set accuracy: {:.3f}".format(clf.score(data_train_normalized,y_train_normalized)))
print("Normalize Verinin Test set accuracy: {:.3f}".format(clf.score(data_test_normalized,y_test_normalized)))
print("Normalize Verinin Test2 set accuracy: {:.3f}".format(clf.score(data_test2_normalized,y_test2_normalized)))
clf.fit(data_train_scaled,y_train_scaled)
print("------")
print("Ölçeklenmiş Verinin Train set accuracy: {:.3f}".format(clf.score(data_train_scaled,y_train_scaled)))
print("Ölçeklenmiş Verinin Test set accuracy: {:.3f}".format(clf.score(data_test_scaled,y_test_scaled)))
print("Ölçeklenmiş Verinin Test2 set accuracy: {:.3f}".format(clf.score(data_test2_scaled,y_test2_scaled)))
```

Orjinal veri, train ile test accuracylaeri arasında fark büyüktür. Bu durumda, ezbere gittiğini ve overfitting olduğunu göstermektedir. Aynı şekilde normalize ve ölçeklendirilmiş veri üzerindeki accuracy değerlerine bakınca overfitting durumu görülmektedir.

#                              LOGISTIC REGRESSION

İkinci kullanılan algoritma Logistic Regression' dır. Bu algoritma lineer modellerin sınıflandırması için kullandığı bir algoritmadır.Lineer modeller doğrusal bir fonksiyon kullanır.
 
                       y=w[0]*x[0]+w[1]*x[1]*w[2]*x[2]+....+b>0
                       
Doğrusal model tahmin formülünün değerinin 0dan büyük olanlar ve 0' dan küçük olanları 2 farklı sınıfa ayırrır.
                       
İsmine rağmen regression algoritması değil, sınıflandırma algoritmasıdır. Regularization yapıyor. Paratme olarak düzleştirme katsayısı olarak C kullanır. Default olarak C, 1 değeri alır.

Önce C normalleştirme katsayısı default değeri ile incelenmiştir.

```python
#LOGISTIC REGRESSION
logreg = LogisticRegression().fit(data_train, y_train)
print("Orjinal verinin Training set score: {:.3f}".format(logreg.score(data_train, y_train)))
print("Orjinal verinin Test set score: {:.3f}".format(logreg.score(data_test, y_test)))
print("Orjinal verinin Test2 set score: {:.3f}".format(logreg.score(data_test2, y_test2)))
print("-------")
logreg = LogisticRegression().fit(data_train_normalized, y_train_normalized)
print("Normalize edilmiş verinin Training set score: {:.3f}".format(logreg.score(data_train_normalized, y_train_normalized)))
print("Normalize edilmiş verininTest set score: {:.3f}".format(logreg.score(data_test_normalized, y_test_normalized)))
print("Normalize edilmiş verinin Test2 set score: {:.3f}".format(logreg.score(data_test2_normalized, y_test2_normalized)))
print("-------")
logreg = LogisticRegression().fit(data_train_scaled, y_train_scaled)
print("Ölçeklendirilmiş Training set score: {:.3f}".format(logreg.score(data_train_scaled, y_train_scaled)))
print("Ölçeklendirilmiş Test set score: {:.3f}".format(logreg.score(data_test_scaled, y_test_scaled)))
print("Ölçeklendirilmiş Test2 set score: {:.3f}".format(logreg.score(data_test2_scaled, y_test2_scaled)))
```

Orjinal verinin test verisiyle arasında çok fark yok gibi ne overfit nede underfit denilebilir. Ama test2 verisi, traine göre daha yüksektir bu durumda test2 için underfit durumu vardır modelin öğrenmesi yetersiz gelmiştir. Ölçeklendirilmiş verinin test verileri için normal gözüküyor, ne underfit ne overfit denilebilir. Ama test2 verileri üzerinde yine underfitting durumundadır. ### RANDOM FOREST


Bir önceki DecisionTree algoritmasında da bahsedildiği gibi DecisionTree algoritmasının hemen overfit olma özelliğine çözüm olarka RandomForest algoritması kullanılmaktadır. Random Forest algoritması birden fazla decision tree'lerin farklı noktalarda, farklı verisetleri üzerinde, farklı featurelarda paralel olarak çalıştırılıp bu ağaçların toplamının ortalaması alınmasına dayanmaktadır. Datayı alıyor, Random olarak örneklendiriyor. Her node, farklı datayla başlıyor, featureları paylaştırıyor.

Parametre olarak maksimum ağaç derinliği girilerek ağacın derinliği sınırlandırılıyor. Parametreside n_estimator, buda kaç ağaç kullanılacağının sayısını veriyor. max_feature parametresi ile kullanması istenilen özellik sayısı belirtilir. max_feature sayısını 1 yaparsak tek bir özellik bakmak zorunda olur, seçeneği olmaz, n özelliğe baksın istersekte tüm özelliklere bakar.
